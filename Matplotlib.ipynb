### Legend 

```python
kamal = [1,2,3,4,5,6,7]


hasan = [50,51,52,48,47,49,46]
Ash = [43,42,40,44,33,35,37]
rohith = [45,48,48,46,40,42,41]



plt.plot(kamal,hasan)
plt.plot(kamal , Ash)
plt.plot(kamal , rohith)
plt.show()
```

```python
kamal = [1,2,3,4,5,6,7]


hasan = [50,51,52,48,47,49,46]
Ash = [43,42,40,44,33,35,37]
rohith = [45,48,48,46,40,42,41]



plt.plot(kamal,hasan , label = 'kamal-hasan-values')
plt.plot(kamal , Ash,label = 'kamal-ash-values')
plt.plot(kamal , rohith , label = 'kamal-rohit-values')
plt.legend(loc = 'best')
plt.show()
```

## Subplots

```python
a = [2014 ,2015 , 2016 , 2017]
b = [4000,4500,5300,4600]
c = [2800,3000,2800,3400]


plt.bar(a,b,color = 'red')
plt.bar(a,c , color = 'yellow')
plt.show()
```

```python
a = [2014 ,2015 , 2016 , 2017]
b = [4000,4500,5300,4600]
c = [2800,3000,2800,3400]

plt.subplot(1,2,1)
plt.bar(a,b,color = 'red')
plt.subplot(1,2,2)
plt.bar(a,c , color = 'yellow')
plt.show()
```

```python
a = [2014 ,2015 , 2016 , 2017]
b = [4000,4500,5300,4600]
c = [2800,3000,2800,3400]

plt.subplot(2,1,1)
plt.bar(a,b,color = 'red')

plt.subplot(2,1,2)
plt.bar(a,c , color = 'yellow')
plt.show()
```

### Pie chart 

```python
x = [40,30,20,10,1]

y = ['Java','Python','c#','PHP','c++']


plt.pie(x , labels=y)
plt.show()
```

```python
plt.figure(figsize = (7,7))

x = [40,30,20,10,1]

y = ['Java','Python','c#','PHP','c++']


plt.pie(x , labels=y)
plt.show()
```

```python
plt.figure(figsize = (7,7))

x = [40,30,20,10,1]

y = ['Java','Python','c#','PHP','c++']


plt.pie(x , labels=y , shadow=True)
plt.show()
```

```python
plt.figure(figsize = (7,7))

x = [40,30,20,10,1]

y = ['Java','Python','c#','PHP','c++']


plt.pie(x , labels=y , shadow=True , autopct='%1.1f%%')
plt.show()
```
